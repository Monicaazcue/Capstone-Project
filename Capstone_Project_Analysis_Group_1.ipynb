```python

import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
data = pd.read_excel('/content/G1.xlsx')
```

```python

# Basic Descriptive Statistics
print(data.describe())


# Distribution of AI Awareness and Interaction
print(data['AI Awareness'].value_counts(normalize=True) * 100)
print(data['AI Interaction'].value_counts(normalize=True) * 100)
```

```python
import matplotlib.pyplot as plt

# Pie charts for binary variables
binary_cols = ['Data Privacy Concerns', 'AI Interaction', 'Preference for Human Interaction']

# Adjust figure size
plt.figure(figsize=(12, 12))

for i, column in enumerate(binary_cols, 1):
    plt.subplot(3, 3, i)
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['tomato', 'lightblue'], fontsize=10)
    plt.title(f'Proportion of {column}', fontsize=12)
    plt.ylabel('')  # Remove y-label for cleaner pie charts

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Setting up the matplotlib figure
plt.figure(figsize=(20, 20))

# Function to add percentage annotations on the bars
def add_percentage_and_count_labels(ax, data):
    for p in ax.patches:
        height = p.get_height()
        percentage = f'{(height/len(data)*100):.1f}%'
        ax.annotate(percentage,
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='center',
                    xytext=(0, 9),
                    textcoords='offset points')

# Bar charts for categorical and binary variables with count on y-axis and percentage on bars
all_categorical_cols = ['Luxury Wellbeing Centers Visited', 'Visit Frequency',
                        'Visit Objectives', 'AI Interaction', 'Data Privacy Concerns',
                        'Preference for Human Interaction']

for i, column in enumerate(all_categorical_cols, 1):
    plt.subplot(4, 3, i)
    counts = data[column].value_counts()
    ax = counts.plot(kind='bar', color='cadetblue', edgecolor='black')
    plt.title(f'Count and Percentage of {column}')
    plt.xlabel(column)
    plt.xticks(rotation=45)
    plt.ylabel('Count')
    add_percentage_and_count_labels(ax, data)

# Adjust layout
plt.tight_layout()
plt.show()

```

```python
# Setting up the matplotlib figure
plt.figure(figsize=(18, 6))

# Bar Chart for Visit Frequency
plt.subplot(1, 3, 1)
visit_freq_counts = data['Visit Frequency'].value_counts()
ax1 = visit_freq_counts.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Visit Frequency')
plt.ylabel('Count')
plt.xlabel('Frequency')
add_percentage_and_count_labels(ax1, data)

# Histogram for AI Experience Enhancement
plt.subplot(1, 3, 2)
ax2 = data['AI Experience Enhancement'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of AI Experience Enhancement')
plt.xlabel('Rating')
plt.ylabel('Frequency')

# Pie Chart for AI Interaction
plt.subplot(1, 3, 3)
ai_interaction_counts = data['AI Interaction'].value_counts()
ai_interaction_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('AI Interaction')
plt.ylabel('')  # Remove y-label for cleaner pie charts

# Adjust layout
plt.tight_layout()
plt.show()
```
